### Exercice 1 : Création et visualisation d'un graphe simple

#### Objectifs :
1. Créer un graphe non orienté simple.
2. Ajouter des nœuds et des arêtes.
3. Visualiser le graphe avec des étiquettes pour les nœuds.

#### Instructions :
1. Utilisez NetworkX pour créer un graphe non orienté.
2. Ajoutez les nœuds `A, B, C, D`.
3. Ajoutez les arêtes `(A, B), (A, C), (B, D), (C, D)`.
4. Visualisez le graphe en utilisant matplotlib, avec des étiquettes pour les nœuds.

### Exercice 4 : Recherche du plus court chemin

#### Objectifs :
1. Créer un graphe pondéré.
2. Utiliser l'algorithme de Dijkstra pour trouver le plus court chemin.
3. Visualiser le graphe avec le chemin le plus court mis en évidence.

#### Instructions :
1. Créez un graphe pondéré avec les nœuds `A, B, C, D, E` et les arêtes pondérées `(A, B, 1), (A, C, 2), (B, C, 1), (B, D, 3), (C, D, 1), (D, E, 2)`.
2. Utilisez l'algorithme de Dijkstra pour trouver le plus court chemin de `A` à `E`.
3. Visualisez le graphe avec le chemin le plus court mis en évidence.

### Exercice 5 : Détection des cycles dans un graphe

#### Objectifs :
1. Créer un graphe.
2. Vérifier s'il contient des cycles.
3. Identifier et afficher tous les cycles.

#### Instructions :
1. Utilisez NetworkX pour créer un graphe.
2. Ajoutez les nœuds `A, B, C, D, E`.
3. Ajoutez les arêtes `(A, B), (B, C), (C, D), (D, A), (B, D), (C, E)`.
4. Utilisez une méthode pour vérifier la présence de cycles.
5. Identifiez et affichez tous les cycles dans le graphe.
